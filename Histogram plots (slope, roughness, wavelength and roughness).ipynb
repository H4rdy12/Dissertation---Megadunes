```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import scipy
from scipy.stats import mannwhitneyu
```

```python
# Read the csv file generated from processing scripts
wvl_df = pd.read_csv("./wvl_df.csv")

amp_df = pd.read_csv('./amp_df.csv')
print("Wavelength dataframe")
print(wvl_df.head())
print("")
print("Amplitude dataframe")
print(amp_df.head())
print(len(amp_df))
```

```python
#plot wvl histograms
fig, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(5,1,figsize=(20,12),tight_layout=True)
ax1.hist(wvl_df['A'],bins = 300,edgecolor='black', facecolor='gray', alpha=0.6) #Add range=[x,y] to remove outliers typical range 2-8m 
ax2.hist(wvl_df['B'],bins = 300,edgecolor='black', facecolor='gray', alpha=0.6)
ax3.hist(wvl_df['C'],bins = 250,edgecolor='black', facecolor='gray', alpha=0.6)
ax4.hist(wvl_df['D'],bins = 150,edgecolor='black', facecolor='gray', alpha=0.6)
ax5.hist(wvl_df['E'],bins = 200,edgecolor='black', facecolor='gray', alpha=0.6)

ax1.set_ylabel('frequency',fontsize=18)
ax2.set_ylabel('frequency',fontsize=18)
ax3.set_ylabel('frequency',fontsize=18)
ax4.set_ylabel('frequency',fontsize=18)
ax5.set_ylabel('frequency',fontsize=18)

ax1.set_title('(A)', loc='left',fontsize=22)
ax2.set_title('(B)', loc='left',fontsize=22)
ax3.set_title('(C)', loc='left',fontsize=22)
ax4.set_title('(D)', loc='left',fontsize=22)
ax5.set_title('(E)', loc='left',fontsize=22)

ax1.xaxis.get_ticklocs(minor=True)
ax1.minorticks_on()
ax2.xaxis.get_ticklocs(minor=True)
ax2.minorticks_on()
ax3.xaxis.get_ticklocs(minor=True)
ax3.minorticks_on()
ax3.xaxis.get_ticklocs(minor=True)
ax3.minorticks_on()
ax4.xaxis.get_ticklocs(minor=True)
ax4.minorticks_on()
ax5.xaxis.get_ticklocs(minor=True)
ax5.minorticks_on()

ax1.set_xlim(0,17000)
ax2.set_xlim(0,17000)
ax3.set_xlim(0,17000)
ax4.set_xlim(0,17000)
ax5.set_xlim(0,17000)

ax5.set_xlabel('wavelength (m)',fontsize=22)
ax1.tick_params(axis='both', which='both', labelsize=16)                                    
ax2.tick_params(axis='both', which='both', labelsize=16) 
ax3.tick_params(axis='both', which='both', labelsize=16) 
ax4.tick_params(axis='both', which='both', labelsize=16) 
ax5.tick_params(axis='both', which='both', labelsize=16) 
#plt.legend(fontsize=22)
plt.savefig('./All Sites extraction wavelengths hist.jpg',dpi=900, transparent=True)
plt.show()
```

```python
# plot combined wvl histogram

all_sites_wvl = []
for i in range(len(wvl_df['A'])):
    a = wvl_df['A'][i]
    all_sites_wvl.append(a)

for i in range(len(wvl_df['B'])):
    a = wvl_df['B'][i]
    all_sites_wvl.append(a)

for i in range(len(wvl_df['C'])):
    a = wvl_df['C'][i]
    all_sites_wvl.append(a)

for i in range(len(wvl_df['D'])):
    a = wvl_df['D'][i]
    all_sites_wvl.append(a)

for i in range(len(wvl_df['E'])):
    a = wvl_df['E'][i]
    all_sites_wvl.append(a)
    
plt.figure(figsize=(20,5))
plt.hist(all_sites_wvl,bins = 200,edgecolor='black', facecolor='steelblue', alpha=0.8)
plt.xlim(0,17000)

plt.xlabel('wavelength (m)',fontsize=22)
plt.ylabel('frequency',fontsize=18)
plt.xticks(fontsize=16) 
plt.yticks(fontsize=16)  
plt.savefig('./Combined Sites wavelengths hist.jpg',dpi=900, transparent=True)
plt.show()
```

```python
# Stats Site comparison
p = mannwhitneyu(wvl_df['A'], wvl_df['B'])
p1 = mannwhitneyu(wvl_df['A'], wvl_df['C'])
p2 = mannwhitneyu(wvl_df['A'], wvl_df['D'])
p3 = mannwhitneyu(wvl_df['A'], wvl_df['E'])

p4 = mannwhitneyu(wvl_df['B'], wvl_df['C'])
p5 = mannwhitneyu(wvl_df['B'], wvl_df['D'])
p6 = mannwhitneyu(wvl_df['B'], wvl_df['E'])

p7 = mannwhitneyu(wvl_df['C'], wvl_df['D'])
p8 = mannwhitneyu(wvl_df['C'], wvl_df['E'])
                                      
p9 = mannwhitneyu(wvl_df['D'], wvl_df['E'])

print("A vs B", p)
print("A vs C",p1)
print("A vs D",p2)
print("A vs E",p3)

print("B vs C",p4)
print("B vs D",p5)
print("B vs E",p6)

print("C vs D",p7)
print("C vs E",p8)

print("D vs E",p9)
```
